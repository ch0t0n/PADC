```python
import os
import re
from tqdm import tqdm
```

```python
def delete_checkpoint_files(root_dir):
    """
    Recursively find and delete files matching:
    checkpoint_epoch{number}.pth
    """

    pattern = re.compile(r"checkpoint_epoch\d+\.pth$")
    files_to_delete = []

    # Progress bar for scanning
    scan_bar = tqdm(desc="Scanning files", unit="file")

    for root, dirs, files in os.walk(root_dir):
        for file in files:
            scan_bar.update(1)

            if pattern.match(file):
                files_to_delete.append(os.path.join(root, file))
                scan_bar.set_postfix(matches=len(files_to_delete), last=file)

    scan_bar.close()

    print(f"\nFound {len(files_to_delete)} checkpoint files.\n")

    # Progress bar for deletion
    delete_bar = tqdm(files_to_delete, desc="Deleting", unit="file")

    for filepath in delete_bar:
        filename = os.path.basename(filepath)
        delete_bar.set_postfix(file=filename)

        try:
            os.remove(filepath)
        except Exception as e:
            print(f"\nError deleting {filepath}: {e}")

    print(f"\nDeleted {len(files_to_delete)} checkpoint files.")
```

```python
parent_folder = r"/home/c/choton/beemachine/codes/AG_vision_2026/"
delete_checkpoint_files(parent_folder)
```
