```python
import torch

ckpt_path = "/home/c/choton/beemachine/everyday_test/sep30_25/lightning_logs/version_5/checkpoints/epoch=199-step=37400.ckpt"
ckpt = torch.load(ckpt_path, map_location="cpu")

print(ckpt.keys())         # see what’s inside
```

```python
state_dict = ckpt["state_dict"]
print(list(state_dict.keys())[:20])  # peek first few keys
```

```python
def guess_model_type(state_dict):
    keys = list(state_dict.keys())
    if any("fpn" in k for k in keys):
        return "FPN"
    elif any("patch_embeddings" in k or "block" in k for k in keys):
        return "SegFormer"
    elif any("decoder.blocks" in k for k in keys):
        return "UNet"
    else:
        return "Unknown"

model_type = guess_model_type(state_dict)
model_type
```
