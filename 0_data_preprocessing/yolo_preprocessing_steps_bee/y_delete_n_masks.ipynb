```python
import os
```

```python
# Paths
dataset_path = "/home/c/choton/beemachine/datasets/Beemachine_Partwhole_v4/"
subset = "train"

images_dir = os.path.join(dataset_path, subset, "aug_images")
masks_dir  = os.path.join(dataset_path, subset, "aug_masks")

# Collect image basenames (without extension)
image_bases = {
    os.path.splitext(f)[0]
    for f in os.listdir(images_dir)
    if f.lower().endswith((".jpg", ".png", ".jpeg"))
}
```

```python
len(os.listdir(images_dir)), len(os.listdir(masks_dir))
```

```python
removed = 0

for mask_file in os.listdir(masks_dir):
    if not mask_file.endswith("_m.png"):
        continue

    # imageName_m.jpg → imageName
    image_base = mask_file[:-6]  # remove "_m.jpg"

    if image_base not in image_bases:
        mask_path = os.path.join(masks_dir, mask_file)
        os.remove(mask_path)
        removed += 1

print(f"✅ Removed {removed} orphan masks")
```
