```python
import os
import shutil
```

```python
# Set the paths
dataset_path = r"/home/c/choton/beemachine/datasets/Beemachine_Partwhole_v4/"
src_images_dir = os.path.join(dataset_path, 'valid', 'images') # r"/home/c/choton/beemachine/datasets/Beemachine_Partwhole_v4/train/images"
src_labels_dir = os.path.join(dataset_path, 'valid', 'labels') # r"/home/c/choton/beemachine/datasets/Beemachine_Partwhole_v4/train/labels"

dst_images_dir = os.path.join(dataset_path, 'test', 'images') # r"/home/c/choton/beemachine/datasets/Beemachine_Partwhole_v4/train/images"
dst_labels_dir = os.path.join(dataset_path, 'test', 'labels') # r"/home/c/choton/beemachine/datasets/Beemachine_Partwhole_v4/train/labels"

# Make sure dstination exists
assert os.path.exists(dst_images_dir), f"The dstination {dst_images_dir} doesn't exist!"
assert os.path.exists(dst_labels_dir), f"The dstination {dst_labels_dir} doesn't exist!"

# Get files only (ignore subfolders) and sort by name
files = sorted(
    [f for f in os.listdir(src_images_dir) if os.path.isfile(os.path.join(src_images_dir, f))]
)

# Take last 10 files
last_10_files = files[-10:]
```

```python
# Move files
for filename in last_10_files:
    src_img_path = os.path.join(src_images_dir, filename)
    dst_img_path = os.path.join(dst_images_dir, filename)

    label_file = os.path.splitext(filename)[0] + ".txt"
    src_label_path = os.path.join(src_labels_dir, label_file)
    dst_label_path = os.path.join(dst_labels_dir, label_file)

    shutil.move(src_img_path, dst_img_path)
    shutil.move(src_label_path, dst_label_path)

    print(f"Moving {filename} and its label...")

print(f"Moved {len(last_10_files)} files.")
```
