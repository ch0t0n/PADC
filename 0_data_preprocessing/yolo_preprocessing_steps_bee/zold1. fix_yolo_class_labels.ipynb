**This code is to fix the incorrect yolo class labels (previously it was 5 with noisy labels)**
Correct class labels: 'abdomen', 'thorax' and 'head'

**Note**: The class labels are already fixed from the roboflow website. No need to fix it anymore!

```python
import os
from tqdm import tqdm

# Class remapping: old → new
remap = {
    2: 0,
    3: 1,
    4: 2
}
```

```python
# Function to fix yolo class labels
def fix_yolo_class_labels(labels_dir):
    label_files = os.listdir(labels_dir)
    for file in tqdm(label_files):
        if not file.endswith(".txt"):
            continue

        file_path = os.path.join(labels_dir, file)
        class_polygons = {}

        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if not parts:
                    continue

                cls = int(parts[0]) # Get the class name
                
                # Skip noise classes
                if cls in [0, 1]:
                    continue

                # Remap class if valid
                if cls in remap:
                    parts[0] = str(remap[cls]) # Get the correct label using the remap dict
                    coords = parts[1:]

                    # Keep the longest polygon per class
                    if parts[0] not in class_polygons or len(coords) > len(class_polygons[parts[0]]):
                        class_polygons[parts[0]] = coords
        
        # Build final lines (one per class, longest polygon)
        new_lines = ["{} {}".format(cls, " ".join(coords)) for cls, coords in class_polygons.items()]

        # Overwrite with cleaned labels
        with open(file_path, "w") as f:
            f.write("\n".join(new_lines))
```

```python
# Path to your YOLO segmentation labels
labels_dir = "/home/c/choton/beemachine/everyday_test/oct1_25/partwhole_yolo_original_fixed/labels" # Fixing test dataset labels
fix_yolo_class_labels(labels_dir)
```

```python
def remove_empty_polygons(labels_dir):
    label_files = os.listdir(labels_dir)
    for file in tqdm(label_files):
        if not file.endswith(".txt"):
            continue

        file_path = os.path.join(labels_dir, file)
        new_lines = []
        modified = False  # track if we removed anything

        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if not parts:
                    continue

                cls = parts[0]
                coords = parts[1:]

                # 🚨 Skip empty polygons
                if len(coords) == 0:
                    modified = True
                    print("Empty polygonal annotations found!")
                    continue

                new_lines.append(" ".join([cls] + coords))

        # Only overwrite if changes were made
        if modified:
            with open(file_path, "w") as f:
                f.write("\n".join(new_lines))
```

```python
# Path to your YOLO segmentation labels
labels_dir = "/home/c/choton/beemachine/everyday_test/oct1_25/partwhole_yolo_original_fixed/labels" # Fixing test dataset labels
remove_empty_polygons(labels_dir)
```
