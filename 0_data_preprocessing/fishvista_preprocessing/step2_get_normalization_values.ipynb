```python
import os
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
```

```python
# Path to CUB images folder (e.g., CUB_200_2011/images)
DATA_DIR = r'/home/c/choton/beemachine/datasets/Others/CUB_200_2011/images'

# Only convert to tensor — no normalization yet
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # keep consistent with your training
    transforms.ToTensor()
])

dataset = datasets.ImageFolder(DATA_DIR, transform=transform)

loader = DataLoader(
    dataset,
    batch_size=64,
    num_workers=4,
    shuffle=False,
    pin_memory=True
)

mean = torch.zeros(3)
std = torch.zeros(3)
total_images = 0

for images, _ in tqdm(loader):
    batch_samples = images.size(0)
    images = images.view(batch_samples, 3, -1)

    mean += images.mean(2).sum(0)
    std += images.std(2).sum(0)
    total_images += batch_samples

mean /= total_images
std /= total_images

print("Mean:", mean)
print("Std:", std)
```
